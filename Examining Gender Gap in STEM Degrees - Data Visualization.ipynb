# Project : Visualizing the gender gap in college degrees

Drawing line charts to visualize gender gap in STEM degrees. Grouping all the degrees in three categories, STEM , liberal arts and other categories.
* STEM - ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
* liberal arts - ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
* other categories - ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

   All three of these lists are ordered in descending order by the percentage of degrees awarded to women


```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Improving the plot asthetics

```python
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees.head()
```

```python
# Removing the x axis labels so that they dont overalp with plot titles and also reduces redundant information
# removing the y axis labels except 0,50 and 100. adding a refernce line at 50%. 

# choosing the color blind friendly colors for the plots.
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_gray = (171/255 , 171/255 , 171/255)


# list of three categories - stem , liberal arts and other.
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(21, 21))

# for loop to plot all the subplots in the STEM category    
for sp in range(0,6):
    if sp == 0:
        pos = sp + 1
    elif sp > 0:
        pos = pos + 3
    ax = fig.add_subplot(6,3,pos)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,50,100])
    ax.axhline(50 , c = cb_gray , alpha = 0.3)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off" , labelbottom = "off")
    
    if sp == 0:
        ax.text(2005, 82, 'Women')
        ax.text(2005, 16, 'Men')
    elif sp == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 10, 'Women')
        ax.tick_params(bottom="off", top="off", left="off", right="off" , labelbottom = "on")
        
# for loop to plot all the subplots in the liberal arts category        
for sp in range(0,5):
    if sp == 0:
        pos = sp + 2
    elif sp > 0:
        pos = pos + 3
    ax = fig.add_subplot(6,3,pos)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,50,100])
    ax.axhline(50 , c = cb_gray , alpha = 0.3)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off" , labelbottom = "off")
    
    if sp == 0:
        ax.text(2005, 20, 'Men')
        ax.text(2003, 80, 'Women')
    if sp == 4:
        ax.tick_params(bottom="off", top="off", left="off", right="off" , labelbottom = "on")
        
# for loop to plot all the subplots in the other category
for sp in range(0,6):
    if sp == 0:
        pos = sp + 3
    elif sp > 0:
        pos = pos + 3
    ax = fig.add_subplot(6,3,pos)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,50,100])
    ax.axhline(50 , c = cb_gray , alpha = 0.3)
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off"  , labelbottom = "off")
    
    if sp == 0:
        ax.text(2005, 8, 'Men')
        ax.text(2003, 90, 'Women')
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(bottom="off", top="off", left="off", right="off"  , labelbottom = "on")

        
        

        
        
plt.savefig("gender_degrees.png")        
plt.show()
```
